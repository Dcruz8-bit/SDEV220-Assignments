```python

# David Cruz Python Assignments Notebook

# Module 2 Chapter 3: Things to Do (3.1 to 3.6)

# 3.1 How many seconds are in an hour?
60 * 60

# 3.2 Assign the result to a variable
seconds_per_hour = 60 * 60
seconds_per_hour

# 3.3 How many seconds are in a day?
seconds_per_hour * 24

# 3.4 Save the result in a variable
seconds_per_day = seconds_per_hour * 24
seconds_per_day

# 3.5 Floating-point division
seconds_per_day / seconds_per_hour

# 3.6 Integer division
seconds_per_day // seconds_per_hour

# Yes, the result of the integer division in 3.6 matches the floating-point result in 3.5,
# except for the `.0` at the end. Both show that there are 24 hours in a day.


# Module 3 Conditionals (Assignments 4.1 & 4.2)

# Assignment 4.1
# This code compares a guessed number to a secret number and prints whether the guess is too low, too high, or just right.

secret = 7
guess = 4

if guess < secret:
    print('too low')
elif guess > secret:
    print('too high')
else:
    print('just right')

# Assignment 4.2
# This code uses boolean variables to determine which fruit or vegetable matches the traits "small" and "green".

small = True
green = True

if small and green:
    print('pea')
elif small and not green:
    print('cherry')
elif not small and green:
    print('watermelon')
else:
    print('pumpkin')


# Module 4 Loops (Assignments 6.1 to 6.3)

# Assignment 6.1
# This code uses a for loop to print the values in the list [3, 2, 1, 0].

for i in [3, 2, 1, 0]:
    print(i)

# Assignment 6.2
# This code uses a while loop to compare a number to guess_me and prints feedback until the correct value is found or exceeded.

guess_me = 7
number = 1

while True:
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
    number += 1

# Assignment 6.3
# This code uses a for loop to compare numbers in a range to guess_me and prints feedback, breaking when the correct value is found or exceeded.

guess_me = 5

for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
```
